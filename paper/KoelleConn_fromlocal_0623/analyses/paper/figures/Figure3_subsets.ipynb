```python
import pandas as pd
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt
```

```python
#sns.heatmap(np.log(output), vmin = -6, vmax = 0)

def fix_pdcsv(csv):
    
    #wt_conn = pd.read_csv('/Users/samsonkoelle/wt_leafsmooth_leafsurface_sumtarget.csv',  header=[0,1])
    csv_rownames = np.asarray(csv.iloc[:,0])
    csv = csv.iloc[:,1:]
    csv.index = csv_rownames
    return(csv)

wt_conn = pd.read_csv('/Users/samsonkoelle/alleninstitute/sambranch/mouse_connectivity_models/paper/connectivities/elsumsurfleafsmth_v3_leafsum_wt.csv', header = [0,1])
wt_conn = fix_pdcsv(wt_conn)

Ntsr1_conn = pd.read_csv('/Users/samsonkoelle/alleninstitute/sambranch/mouse_connectivity_models/paper/connectivities/elsumsurfleafsmth_v3_leafsum_Ntsr1-Cre_GN220.csv', header = [0,1])
Ntsr1_conn = fix_pdcsv(Ntsr1_conn)
Cux2_conn = pd.read_csv('/Users/samsonkoelle/alleninstitute/sambranch/mouse_connectivity_models/paper/connectivities/elsumsurfleafsmth_v3_leafsum_Cux2-IRES-Cre.csv', header = [0,1])
Cux2_conn = fix_pdcsv(Cux2_conn)
Rbp4_conn = pd.read_csv('/Users/samsonkoelle/alleninstitute/sambranch/mouse_connectivity_models/paper/connectivities/elsumsurfleafsmth_v3_leafsum_Rbp4-Cre_KL100.csv', header = [0,1])
Rbp4_conn = fix_pdcsv(Rbp4_conn)                
Tlx3_conn = pd.read_csv('/Users/samsonkoelle/alleninstitute/sambranch/mouse_connectivity_models/paper/connectivities/elsumsurfleafsmth_v3_leafsum_Tlx3-Cre_PL56.csv', header = [0,1])
Tlx3_conn = fix_pdcsv(Tlx3_conn)
```

```python
source_regions = np.asarray(['VISp2/3', 'VISp4', 'VISp5', 'VISp6a','VISl4', 'VISl5', 'VISl6a'])
target_regions = [(  'ipsi',     'LGv'), (  'ipsi',     'LGd'),(  'ipsi',     'VISp'), (  'ipsi',     'VISl')]
targ_reg = [(  'ipsi',     'ORBvl'), (  'ipsi',     'AId'),(  'ipsi',     'AIv'), (  'ipsi',     'TEa'),
            (  'ipsi',     'ECT'), (  'ipsi',     'BLA'),(  'ipsi',     'CEA')]
source_reg = np.asarray(['MOp2/3', 'MOp5', 'MOp6a','MOs2/3', 'MOs5', 'MOs6a' ])
```
