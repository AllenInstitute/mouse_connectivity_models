```python
import pandas as pd
import os
from tqdm import tqdm
import numpy as np
import sys
import os

def fix_pdcsv(csv):
    
    #wt_conn = pd.read_csv('/Users/samsonkoelle/wt_leafsmooth_leafsurface_sumtarget.csv',  header=[0,1])
    csv_rownames = np.asarray(csv.iloc[:,0])
    csv = csv.iloc[:,1:]
    csv.index = csv_rownames
    return(csv)

def fill_df_na(data, indices):
    
    for i in range(indices.shape[0]):
        data.iloc[indices[i,0],indices[i,1]] = np.nan
        
    return(data)

def load_csv(path):
    bad_csv = pd.read_csv(path, header = [0,1])
    good_csv = fix_pdcsv(bad_csv)
    return(good_csv)
```

```python
working_directory = os.popen('git rev-parse --show-toplevel').read()[:-1]
sys.path.append(workingdirectory)
os.chdir(workingdirectory)
path_thresholded = "paper/connectivities/revision_thresholded"
path_unthresholded = 'paper/connectivities/revision_unthresholded'
connectivities_unthresholded = os.listdir(f"{working_directory}/{path_unthresholded}")
```

```python
thresh = 1e-6
for connectivity in connectivities_unthresholded:
    print(connectivity)
    data = load_csv(f"{working_directory}/{path_unthresholded}/{connectivity}")
    data[data< thresh] = 0.
    data.to_csv(f"{working_directory}/{path_thresholded}/{connectivity}")
```
