Norms proceed as
\begin{eqnarray*}
y_j = \sum w_{ij} y_i
\end{eqnarray*}
Then
\begin{eqnarray*}
\|y_j\| &= \|\sum w_{ij} y_i\|
&= (\sum_t (\sum_i w_{ij} y_{it} ) ^2)^{1/2} \\
&=  (\sum_t (w_{j}^T y_{t} ) ^2)^{1/2}
\end{eqnarray*}



\begin{eqnarray*}
y_j = ( \sum w_{ij} ( y_i^{\odot2} ))^{\odot 1/2}
\end{eqnarray*}
Then
\begin{eqnarray*}
\|y_j\| &=
\|(  \sum_i w_{ij} ( y_i^{\odot2} ))^{\odot 1/2}\| \\
&= (\sum_{t} \sum_i w_{ij} ( y_{it}^{2} )) )^{1/2} \\
&= ( \sum_i w_{ij} \sum_{t}  ( y_{it}^2 )) )^{1/2} \\
&= ( \sum_i w_{ij} \| y_{i}\|^2 )) ^{1/2} \\
&= 1
\end{eqnarray*}
